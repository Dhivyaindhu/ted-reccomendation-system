```python
%%capture
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import nltk
import string
import warnings
from scipy.stats import pearsonr
from nltk.corpus import stopwords
from wordcloud import WordCloud
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/tedx_dataset.csv')
print(df.head())
```

```python
splitted = df['posted'].str.split(' ', expand=True)

# Creating columns for month and year of the talk
df['year'] = splitted[2].astype('int')
df['month'] = splitted[1]
```

```python

# Let's combine the title and the details of the talk.
df['details'] = df['title'] + ' ' + df['details']

# Removing the unnecessary information
df = df[['main_speaker', 'details']]
df.dropna(inplace = True)
df.head()
```

```python
# We would like to have a copy of our data for future use.
data = df.copy()
```

```python
def remove_stopwords(text):
  stop_words = stopwords.words('english')

  imp_words = []

  # Storing the important words
  for word in str(text).split():
    word = word.lower()

    if word not in stop_words:
      imp_words.append(word)

  output = " ".join(imp_words)

  return output
```

```python
df['details'] = df['details'].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
punctuations_list = string.punctuation


def cleaning_punctuations(text):
    signal = str.maketrans('', '', punctuations_list)
    return text.translate(signal)


df['details'] = df['details'].apply(lambda x: cleaning_punctuations(x))
df.head()
```

```python
from sentence_transformers import SentenceTransformer

# Initialize the pre-trained Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')  # Lightweight and efficient

# Assuming 'details' is the column in your DataFrame containing text data
corpus = df['details'].tolist()  # Convert the 'details' column into a list of sentences

# Generate embeddings for the text data
embeddings = model.encode(corpus, show_progress_bar=True)

# The embeddings are now ready to be used for similarity computations or further processing
print(f"Generated {len(embeddings)} sentence embeddings with dimension {len(embeddings[0])}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr

def get_similarities(talk_content, data=df, model=None):
    """
    Compute cosine similarities and Pearson correlations for a given talk_content
    against all items in the dataset using Sentence-BERT embeddings.

    Parameters:
    - talk_content: str, the talk content to compare.
    - data: DataFrame, dataset with a 'details' column.
    - model: SentenceTransformer, pre-trained Sentence-BERT model instance.

    Returns:
    - sim: list, cosine similarities for each row.
    - pea: list, Pearson correlations for each row.
    """
    if model is None:
        raise ValueError("Please provide a pre-trained Sentence-BERT model instance.")

    # Generate embedding for the input talk_content
    talk_embedding = model.encode([talk_content])

    # Pre-compute embeddings for the dataset if not already done
    if 'embeddings' not in data:
        data['embeddings'] = data['details'].apply(lambda x: model.encode([x])[0])

    # Store similarities
    sim = []
    pea = []

    for idx, row in data.iterrows():
        # Get embedding for the current 'details'
        current_embedding = row['embeddings']

        # Compute cosine similarity
        cos_sim = cosine_similarity(talk_embedding, [current_embedding])[0][0]

        # Compute Pearson correlation
        pea_sim = pearsonr(talk_embedding.squeeze(), current_embedding.squeeze())[0]

        sim.append(cos_sim)
        pea.append(pea_sim)

    return sim, pea
```

```python
def recommend_talks(talk_content, data=df, model=None):
    """
    Recommend talks similar to the given talk_content based on Sentence-BERT embeddings.

    Parameters:
    - talk_content: str, the talk content to compare.
    - data: DataFrame, dataset with a 'details' column and precomputed embeddings.
    - model: SentenceTransformer, pre-trained Sentence-BERT model instance.

    Returns:
    - None, but displays the top 5 most similar talks.
    """
    if model is None:
        raise ValueError("Please provide a pre-trained Sentence-BERT model instance.")

    # Compute cosine similarities and Pearson correlations
    data['cos_sim'], data['pea_sim'] = get_similarities(talk_content, data=data, model=model)

    # Sort by cosine similarity and Pearson correlation
    data.sort_values(by=['cos_sim', 'pea_sim'], ascending=[False, False], inplace=True)

    # Display the top 5 most similar talks
    print("\nTop 5 Recommended Talks:")
    display(data[['main_speaker', 'details', 'cos_sim', 'pea_sim']].head(5))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr

# Initialize the Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to compute similarities using embeddings
def get_similarities(talk_content, data, model):
    # Getting vector for the input talk_content (ensure it's 2D)
    talk_content_embedding = model.encode(talk_content)

    # We will store similarity for each row of the dataset
    sim = []
    pea = []

    for idx, row in data.iterrows():
        details = row['details']

        # Getting vector for current talk's details (ensure it's 2D)
        talk_details_embedding = model.encode(details)

        # Reshape both embeddings to 2D for cosine similarity (1, n)
        talk_content_embedding_reshaped = talk_content_embedding.reshape(1, -1)
        talk_details_embedding_reshaped = talk_details_embedding.reshape(1, -1)

        # Calculating cosine similarity
        cos_sim = cosine_similarity(talk_content_embedding_reshaped, talk_details_embedding_reshaped)[0][0]

        # Calculating Pearson correlation
        pea_sim = pearsonr(talk_content_embedding.squeeze(), talk_details_embedding.squeeze())[0]

        sim.append(cos_sim)
        pea.append(pea_sim)

    return sim, pea

# Function to recommend talks
def recommend_talks(talk_content, data, model):
    # Compute cosine similarities and Pearson correlations
    data['cos_sim'], data['pea_sim'] = get_similarities(talk_content, data=data, model=model)

    # Sort by cosine similarity and Pearson correlation
    data.sort_values(by=['cos_sim', 'pea_sim'], ascending=[False, False], inplace=True)

    # Display the top 5 most similar talks
    print("\nTop 5 Recommended Talks:")
    print(data[['main_speaker', 'details', 'cos_sim', 'pea_sim']].head(5))

# Input talk content as a single-element list
talk_content = ['Climate change and impact on the health. How can we change this world by reducing carbon footprints?']

# Recommend similar talks
recommend_talks(talk_content, data=df, model=model)
```
